Load and organise intrinsic features

```python
EG1_seqinr = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/seqinr/EG_1_seqinr.csv')
EG2_seqinr = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/seqinr/EG_2_seqinr.csv')

NEG1_seqinr = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/seqinr/NEG_1_seqinr.csv')
NEG2_seqinr = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/seqinr/NEG_2_seqinr.csv')
NEG3_seqinr = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/seqinr/NEG_3_seqinr.csv')
NEG4_seqinr = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/seqinr/NEG_4_seqinr.csv')

# Merge all rows together (total 5845 rows)
seqinr_matrix = pd.concat([
    EG1_seqinr,
    EG2_seqinr,
    NEG1_seqinr,
    NEG2_seqinr,
    NEG3_seqinr,
    NEG4_seqinr
], ignore_index=True)

# Drop peptide ID column
seqinr_matrix = seqinr_matrix.drop('ensembl_peptide_id', axis=1)
```

```python
# Load rDNAse features and add labels (0 for non-essential, 1 for essential)
EG1_rdnase = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/rdnase/EG_1_rdnase.csv')
EG2_rdnase = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/rdnase/EG_2_rdnase.csv')

NEG1_rdnase = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/rdnase/NEG_1_rdnase.csv')
NEG2_rdnase = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/rdnase/NEG_2_rdnase.csv')
NEG3_rdnase = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/rdnase/NEG_3_rdnase.csv')
NEG4_rdnase = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/rdnase/NEG_4_rdnase.csv')

# Merge all rows together (total 5845 rows)
rdnase_matrix = pd.concat([
    EG1_rdnase,
    EG2_rdnase,
    NEG1_rdnase,
    NEG2_rdnase,
    NEG3_rdnase,
    NEG4_rdnase
], ignore_index=True)

# Drop peptide ID column
rdnase_matrix = rdnase_matrix.drop('ensembl_peptide_id', axis=1)
```

```python
rdnase_matrix.head()
```

```python
# Load protr features and concatenate
EG1_protr = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/protr/EG_1_protr.csv')
EG2_protr = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/protr/EG_2_protr.csv')

NEG1_protr = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/protr/NEG_1_protr.csv')
NEG2_protr = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/protr/NEG_2_protr.csv')
NEG3_protr = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/protr/NEG_3_protr.csv')
NEG4_protr = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/protr/NEG_4_protr.csv')

# Merge all rows together (total 5845 rows)
protr_matrix = pd.concat([
    EG1_protr,
    EG2_protr,
    NEG1_protr,
    NEG2_protr,
    NEG3_protr,
    NEG4_protr
], ignore_index=True)

# Drop peptide ID column
protr_matrix = protr_matrix.drop('ensembl_peptide_id', axis=1)
```

```python
protr_matrix.head()
```

```python
EG1_structure = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/structure/EG_1_structure_features.csv')
EG2_structure = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/structure/EG_2_structure_features.csv')

NEG1_structure = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/structure/NEG_1_structure_features.csv')
NEG2_structure = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/structure/NEG_2_structure_features.csv')
NEG3_structure = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/structure/NEG_3_structure_features.csv')
NEG4_structure = pd.read_csv('/content/drive/My Drive/ICL/FYP/Data/intrinsic_features/structure/NEG_4_structure_features.csv')

# Merge all rows together (total 5845 rows)
structure_matrix = pd.concat([
    EG1_structure,
    EG2_structure,
    NEG1_structure,
    NEG2_structure,
    NEG3_structure,
    NEG4_structure
], ignore_index=True)

# Drop peptide ID column
structure_matrix = structure_matrix.drop('ensembl_peptide_id', axis=1)

# Rename gene ID column
structure_matrix = structure_matrix.rename(columns={'mode_ensembl_gene_id': 'FBgn_ID'})

# Reorder columns to bring FBgn_id to first column
cols = ['FBgn_ID'] + [col for col in structure_matrix.columns if col != 'FBgn_ID']
structure_matrix = structure_matrix[cols]
```

```python
structure_matrix
```

Merge by gene ID

```python
# Merge by each df and check gene count in each step
intrinsic_df = seqinr_matrix.merge(rdnase_matrix, on="FBgn_ID", how="inner")
intrinsic_df # n = 5845 rows
```

```python
intrinsic_df = intrinsic_df.merge(protr_matrix, on="FBgn_ID", how="inner")
```

```python
# structure_matrix does not have data for a large proportion of genes, so need to create a padded matrix with NaNs

all_gene_ids = intrinsic_df['FBgn_ID']
missing_ids = set(all_gene_ids) - set(structure_matrix['FBgn_ID'])

# Get column names (excluding Fbgn_ID)
feature_cols = structure_matrix.columns.drop('FBgn_ID')

# Create rows of zeros for the missing IDs
padding_df = pd.DataFrame(np.nan, index=range(len(missing_ids)), columns=feature_cols)
padding_df['FBgn_ID'] = list(missing_ids)

# Append these padded rows to the structure_matrix
structure_matrix_padded = pd.concat([structure_matrix, padding_df], ignore_index=True)
structure_matrix_padded = structure_matrix_padded.sort_values('FBgn_ID').reset_index(drop=True)
```

```python
structure_matrix_padded
```

```python
print(structure_matrix_padded.dtypes)
```

```python
# Merge padded structure matrix with remaining data
intrinsic_df = intrinsic_df.merge(structure_matrix_padded, on="FBgn_ID", how="left")
```

```python
# Load essential/nonessential gene IDs
essential_genes = pd.read_csv("/content/drive/My Drive/ICL/FYP/Data/labels/essential_genes.txt", header=None)[0].tolist()
nonessential_genes = pd.read_csv("/content/drive/My Drive/ICL/FYP/Data/labels/nonessential_genes.txt", header=None)[0].tolist()
```

```python
# Strip whitespace and ensure same case
intrinsic_df["FBgn_ID"] = intrinsic_df["FBgn_ID"].str.strip()
essential_genes = [g.strip() for g in essential_genes]
nonessential_genes = [g.strip() for g in nonessential_genes]
```

```python
# Assign label
def get_label(gene_id):
    if gene_id in essential_genes:
        return 1
    elif gene_id in nonessential_genes:
        return 0
    else:
        return None

intrinsic_df["label"] = intrinsic_df["FBgn_ID"].apply(get_label)
```

```python
# Numbers should be in line with essentiality annotations (2081 essential, 3764 non-essential genes )
print(intrinsic_df["label"].value_counts(dropna=False))
```
